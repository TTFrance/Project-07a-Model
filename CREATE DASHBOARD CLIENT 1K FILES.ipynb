# NEW 1K PICKLE - 900 GOOD, 100 BAD

```python
dfgood = df[df.TARGET==0].sample(n=900)
dfgood.shape

dfbad = df[df.TARGET==1].sample(n=100)
dfbad.shape

df_manual = df.reset_index()
df_manual = df_manual[df_manual.SK_ID_CURR.isin([226060, 310013, 324973, 449031, 328692, 392503])]
df_manual = df_manual.set_index('SK_ID_CURR')
df_manual.shape

df_new = pd.concat([df_manual,dfgood,dfbad])
df_new.TARGET.value_counts()

df_new.to_pickle('../data/cleaned/client_data_api_dashboard_1k_900g_100b.pkl',protocol=4)
```

# NEW 1K PICKLE - 500 GOOD, 500 BAD

```python
dfgood = df[df.TARGET==0].sample(n=500)
dfgood.shape

dfbad = df[df.TARGET==1].sample(n=500)
dfbad.shape

df_manual = df.reset_index()
df_manual = df_manual[df_manual.SK_ID_CURR.isin([226060, 310013, 324973, 449031, 328692, 392503])]
df_manual = df_manual.set_index('SK_ID_CURR')
df_manual.shape

df_new = pd.concat([df_manual,dfgood,dfbad])
df_new.TARGET.value_counts()

df_new.to_pickle('../data/cleaned/client_data_api_dashboard_1k_500g_500b.pkl',protocol=4)
```
